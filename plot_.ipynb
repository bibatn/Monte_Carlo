```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
number_mpi_processes = [1, 4, 16, 32]
time_1 = [0.00219904, 0.00137322, 0.00084548, 0.00063389]
time_2 = [0.0038484, 0.0018367, 0.0012929, 0.0009872]
time_3 = [0.00391767, 0.00273235, 0.00187639, 0.00127639]
```

```python
s_1 = [1, time_1[0]/time_1[1], time_1[0]/time_1[2], time_1[0]/time_1[3]]
print(s_1)
```

```python
s_2 = [1, time_2[0]/time_2[1], time_2[0]/time_2[2], time_2[0]/time_2[3]]
print(s_2)
```

```python
s_3 = [1, time_3[0]/time_3[1], time_3[0]/time_3[2], time_3[0]/time_3[3]]
print(s_3)
```

```python
tol_1 = [2.19904e-05, 1.37322e-05, 1.27422e-05, 1.27422e-05]
tol_2 = [3.66476e-07, 3.45779e-07, 3.27479e-07, 3.27479e-07]
tol_3 = [3.27479e-07, 3.28379e-07, 3.12478e-07, 3.12478e-07]
```

```python
plt.figure(figsize=[12,8] )
plt.subplot(2,1,1)
plt.plot(number_mpi_processes, time_1, label = 'eps = 0.00003')
plt.plot(number_mpi_processes, time_2, label = 'eps = 0.000005')
plt.plot(number_mpi_processes, time_3, label = 'eps = 0.0000015')
plt.xlabel('Количество MPI процессов', fontsize=16)
plt.ylabel('Времы выполнения', fontsize=16)
plt.legend(fontsize = 16)
plt.grid('on')
plt.subplot(2,1,2)
plt.plot(number_mpi_processes, tol_1, label = 'eps = 0.00003')
plt.plot(number_mpi_processes, tol_2, label = 'eps = 0.000005')
plt.plot(number_mpi_processes, tol_3, label = 'eps = 0.0000015')
plt.xlabel('Количество MPI процессов', fontsize=16)
plt.ylabel('Ошибка', fontsize=16)
plt.legend(fontsize = 16)
plt.grid('on')
plt.savefig('polus.eps', format = 'eps')
plt.show()
```
